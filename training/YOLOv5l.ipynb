# İlk Eğitim (mAP: 0.533)

```python
from ultralytics import YOLO

model = YOLO("yolov5l.pt")

model.train(
    data="VisDrone.yaml",
    epochs=40,
    imgsz=896,
    batch=2,
    device="cuda",
    workers=2,
    cache=False,
    amp=True,
    optimizer="AdamW",
    lr0=0.001,
    close_mosaic=10,
    patience=20,
    plots=True,
    project="/content/drive/MyDrive/yolo_runs",
    name="visdrone_v5l_safe"
)
```

# Finetune (mAP 0.597)

```python
from ultralytics import YOLO

model = YOLO(
    "/content/drive/MyDrive/yolo_runs/visdrone_v5l_safe/weights/best.pt"
)

model.train(
    data="VisDrone.yaml",
    epochs=50,
    imgsz=1280,
    batch=1,
    device="cuda",
    workers=2,
    optimizer="AdamW",
    lr0=0.0003,
    cos_lr=True,
    close_mosaic=0,
    patience=20,
    cache=False,
    amp=True,
    plots=True
)
```
